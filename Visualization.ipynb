###### Ploting two or more line

```python
x = np.linspace(1,15,100)
fig, ax = plt.subplots() # Create a figure and an axes
ax.plot(x,x,label = 'linear')
ax.plot(x,x**2,label = 'quadratic')
ax.plot(x,5*x,label = 'non linear')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('Intioal plot')
ax.legend()
```

```python
y = np.linspace(1,78,100)
plt.plot(y,y,label ='linear')
plt.plot(y,y*10,label = 'cubic')
plt.plot(y,y*4,label = 'fourth power')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Initial plot')
plt.legend()
```

###### Histogram

```python

pop = list(np.random.rand(1,20)*1000)[0]
ind = [i for i in pop]
plt.bar(ind,pop,width = 5)
plt.xlabel('y')
plt.ylabel('x')
plt.title('Bar ')
plt.legend()
plt.show()

```

```python

pop = list(np.random.rand(1,20)*100)[0]
ind = [10,20,30,40,50,60,70,80,90]
plt.hist(pop,ind,histtype = 'bar',rwidth = 0.5)
plt.xlabel('y')
plt.ylabel('x')
plt.title('histgram ')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt  
person_ages = [2,5,70,40,30,45,50,45,43,40,44, 
        60,7,13,57,18,90,77,32,21,20,40] 
  
# setting the ranges and no. of intervals 
range = (0, 100) 
no_of_peole = 10  
  
# plotting a histogram 
plt.hist(person_ages, no_of_peole, range, color = 'green', 
        histtype = 'bar', rwidth = 0.8) 
  
# x-axis label 
plt.xlabel('age') 
# frequency label 
plt.ylabel('No. of people') 
# plot title 
plt.title('histogram') 
  
# function to show the plot 
plt.show() 
```

###### Pie chart

```python
import matplotlib.pyplot as plt 

activities = ['eat', 'sleep', 'work', 'play'] 
slices = [3, 7, 8, 6] 
  
colors = ['r', 'y', 'g', 'b'] 
  
plt.pie(slices, labels = activities, colors=colors,  startangle=90, shadow = True, explode = (0, 0, 0.1, 0), radius = 1.2, autopct = '%1.1f%%') 
  
plt.legend() 
  
plt.show()
```

```python
day = [1,2,3,4,5]
study = [1,5,2,8,7]
eating = [2,10,7,9,3]
sleeping = [8,3,1,14,9]
plt.stackplot(day,study,eating,sleeping,color = ['g','r','b'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('stackplot plot ')
plt.legend()
plt.show()
```

```python

#ploting a graph that show the no.of patient based on there age

data_age = data.groupby(['age']).is_patient.sum()
plt.plot( data_age.index,data_age.values)
plt.xlabel('Age')
plt.ylabel('No. Of People')
plt.title('No. of patient age wise')
plt.show()



data_gen = data.groupby(['gender']).gender.count()
plt.pie(data_gen.values,labels = data_gen.index,colors = ['r','g'],shadow=True,startangle=90,radius = 1.2, autopct = '%1.1f%%')
plt.title('Patient Gender wise')
plt.legend()
plt.show()
```

```python

#ploting a graph that show the no.of patient based on there age
plt.figure(figsize=(40,15))
data_age = data.groupby(['age']).is_patient.sum()
plt.plot( data_age.index,data_age.values)
plt.xlabel('Age')
plt.ylabel('No. Of People')
plt.title('No. of patient age wise')
plt.show()


```

```python
plt.figure(figsize=(30,15))
plt.bar( data_age.index,data_age.values,tick_label = data_age.index,width  =0.5,color= ['red','blue'])
plt.xlabel('Age')
plt.ylabel('No. Of People')
plt.title('No. of patient age wise')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(30,15))
data_male = data[data['gender'] == 'Male'].groupby(['age']).is_patient.sum()
data_female = data[data['gender'] == 'Female'].groupby(['age']).is_patient.sum()
plt.plot( data_male.index,data_male.values,label = 'Male')
plt.plot( data_female.index,data_female.values,label = 'Female')
plt.xlabel('Age')
plt.ylabel('No. Of People')
plt.title('No. of patient age and Gender wise')
plt.legend()
plt.show()
```
